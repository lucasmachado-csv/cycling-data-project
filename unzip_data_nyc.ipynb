```python
import os
import zipfile
```

```python
# Manually downloaded 6 ZIP files and imported them into data_nyc folder 

def unzip_all_recursive(start_dir):
    # Start with a list of all files currently in the directory
    queue = []
    for root, dirs, files in os.walk(start_dir):
        for f in files:
            queue.append(os.path.join(root, f))
    
    # Process the queue until it's empty
    while queue:
        current_file = queue.pop(0) # Get the next file to check
        
        if current_file.lower().endswith(".zip"):
            print(f"Unzipping: {current_file}")
            parent_dir = os.path.dirname(current_file)
            
            try:
                # 1. Unzip
                with zipfile.ZipFile(current_file, 'r') as z:
                    z.extractall(parent_dir)
                    # 2. Add NEW files to the queue to be checked
                    for new_file in z.namelist():
                        full_path = os.path.join(parent_dir, new_file)
                        queue.append(full_path)
                
                # 3. Delete the old zip
                os.remove(current_file)
                
            except Exception as e:
                print(f"Error on {current_file}: {e}")

if __name__ == "__main__":
    unzip_all_recursive("data_nyc")
```
